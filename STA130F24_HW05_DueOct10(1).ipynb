### 3. The second "Pre-lecture" video (above) explains that we "imagine a world where the null hypothesis is true" when calculating a p-value? Explain why this is in your own words in a way that makes the most sense to you.<br>

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>

Imagine the null hypothesis as the "default" assumption that there is no effect or no difference. When we calculate a p-value, we are assuming that this default is true, even though we want to test if there's something unusual in our data.

The **sampling distribution** comes into play here. It represents the range of possible outcomes for our test statistic (like a mean or proportion) under the assumption that the null hypothesis is true. This distribution shows us the typical values we expect to see if everything is just random, with no real effect.

When we calculate a p-value, we're measuring how extreme our actual data is compared to this distribution. A small p-value means our data is far from what we'd expect under the null hypothesis, suggesting something unusual is going on. A large p-value means our data is consistent with the null hypothesis, so we wouldn't reject it.

In short, the idea is: "Imagine the null hypothesis is true, and then check how likely it is to see data as extreme as ours." The sampling distribution is crucial because it tells us what "extreme" or "normal" looks like in this imagined world.

In hypothesis testing, we start with a null hypothesis (H0), which says there is no effect or difference. We collect data and do a statistical test to see if we have enough evidence to reject this hypothesis for an alternative hypothesis (Ha), which suggests there is an effect.

A p-value tells us how likely it is to see our results (or something even more surprising) if the null hypothesis is true. It shows the probability of getting our data under the assumption that nothing is happening.

A small p-value (usually less than 0.05) suggests that the results we observed are very unlikely to occur if we assume that nothing is happening (this assumption is known as the null hypothesis, which is a statement that there is no effect or difference). For example, a p-value of 0.01 means there's only a 1% chance we would see such unusual results just by random chance if the null hypothesis is correct. This small p-value comes from comparing our actual results (referred to as the observed test statistic, which summarizes how far our results are from what we would expect if the null hypothesis were true) to what we would expect to see if nothing were really going on (this is represented by the sampling distribution, which shows how a test statistic would behave under the null hypothesis). When the p-value is low, it gives strong evidence against the null hypothesis, meaning that our results are so surprising that we start to doubt the idea that nothing is happening. Ultimately, a small p-value makes it seem "ridiculous" to believe theres no effect or difference when our findings contradict that idea.

# problem statement

In the original experiment, Ronald Fisher tested whether Dr. Muriel Bristol could reliably taste the difference between cups of tea where milk was added first and cups where tea was added first. Fisher proposed that any correct identification by Bristol might be due to chance, and thus conducted a hypothesis test. In this STA130 experiment, the setup is similar, but the population (students) and the sample size differ. In our experiment we have 80 students (sample size) and in the Fisher's tea experiment the sample size is 1 (Dr. Bristol). The context is broader and perhaps less personalized than the original study. We aim to investigate whether students in the STA130 class can accurately identify the order in which milk and tea are poured into a cup. Specifically, we hypothesize that the students' ability to distinguish between the two pouring methods is better than what would be expected by random guessing. To test this hypothesis, we conducted a study involving a random sample of 80 students, out of which 49 correctly identified whether the milk or tea was poured first. We will use bootstrapping simulations to analyze the data and determine if the observed proportion of correct identifications significantly differs from the null hypothesis, which states that there is no difference in performance (i.e., students are guessing with a 50% success rate). By assessing the results of this test, we aim to draw conclusions about the students' perceptual abilities regarding the order of pouring and whether these abilities reflect a genuine distinction between the two methods. 

# hypotheses: 
**formal null hypothesis**: The STA130 students cannot taste the difference between tea being poured first vs milk being poured first. 

**informal null hypothesis**: We believe that the students can't tell whether the tea or milk was poured first, they're just guessing randomly. Since there are only two options (milk first or tea first), we expect about half of them to guess correctly just by chance. 
The students are just guessing whether the milk or tea was poured first, so about half of them will get it right by chance.

**formal null hypothesis based on population parameter**: **h0=0.5**
We assume that the true proportion of students who can accurately identify the pouring order is no better than 50%, which aligns with random guessing. Since students can either say that the tea was poured first or the milk was poured first, their chances of being correct would be equal, resulting in a 50-50 probability. Side note: this is the value we will compare our observed proportion (49/80) against. We will use the calculated p-value to quantify our conclusion based on the observed proportion in relation to the null hypothesis, which assesses the probability of observing data as extreme as 49 correct identifications or more, assuming the true success rate is 50%. By comparing this p-value to the significance level ( = 0.05), we can determine whether to reject the null hypothesis (H).

**alternative hypothesis**:
The STA130 students can accurately taste the difference and have a success rate higher than random guessing.

# quantitative analysis

Each students ability to correctly identify whether the milk or tea was poured first is considered an independent event. Since we have a fixed number of **independent** trials, in this case 80 students, we can use bootstrapping to determine whether our observed proportion (the success rate of determining milk first vs tea first) significantly differs from what we would expect by chance. We are able to assess the likelihood of observing the number of correct identifications by chance alone based on hypothesis testing. This helps us determine if there is a significant effect related to the pouring order of the tea. From this, we can determine whether this provides enough evidence for us to reject our null hypothesis, which is our goal. We'll then calculate the p-value to assess the evidence against H0. If the p-value is sufficiently small (e.g., less than 0.05), we will reject the null hypothesis in favor of the alternative.

# code
using bootstrapping to do hypothesis testing, then use binomial testing (recommended by ChatGPT even though Scott hasn't taught it) to double check our p-value

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
n = 80  # sample size
observed_successes = 49  # observed number of successes
observed_proportion = observed_successes / n
p0 = 0.5  # null hypothesis proportion
n_simulations = 10000  # number of bootstrap samples

# Bootstrap resampling under H
np.random.seed(42)
bootstrap_proportions = np.random.binomial(n=n, p=p0, size=n_simulations) / n

# Calculate p-value (proportion of samples greater than or equal to observed proportion)
p_value = np.sum(bootstrap_proportions >= observed_proportion) / n_simulations

# KDE plot with confidence intervals
sns.kdeplot(bootstrap_proportions, shade=True, color="lightblue")
plt.axvline(x=observed_proportion, color='red', linestyle='--', label=f'Observed Proportion ({observed_proportion:.3f})')
plt.axvline(x=p0, color='green', linestyle='--', label='Null Hypothesis Proportion (0.5)')

# Confidence interval (e.g., 95%)
ci_lower, ci_upper = np.percentile(bootstrap_proportions, [2.5, 97.5])
plt.axvline(x=ci_lower, color='yellow', linestyle='-', label=f'95% CI Lower ({ci_lower:.3f})')
plt.axvline(x=ci_upper, color='yellow', linestyle='-', label=f'95% CI Upper ({ci_upper:.3f})')

# Plot settings
plt.xlabel('Proportion of Correct Identifications')
plt.ylabel('Density')
plt.title(f'Bootstrap Distribution of Proportions under H (p = {p0})\nP-value: {p_value:.4f}')
plt.legend()
plt.show()

print(f"P-value from bootstrapping: {p_value:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
n = 80  # Total number of students
observed_successes = 49  # Number of correct identifications
num_bootstrap_samples = 1000  # Number of bootstrap samples

# Original sample (binary data, 1 for correct identification, 0 for incorrect)
original_sample = np.array([1]*observed_successes + [0]*(n - observed_successes))

# Bootstrapping
bootstrap_proportions = []
np.random.seed(0)  # For reproducibility
for _ in range(num_bootstrap_samples):
    bootstrap_sample = np.random.choice(original_sample, size=n, replace=True)
    bootstrap_proportion = np.mean(bootstrap_sample)
    bootstrap_proportions.append(bootstrap_proportion)

# Calculate confidence intervals
confidence_level = 0.95
lower_bound = np.percentile(bootstrap_proportions, (1 - confidence_level) / 2 * 100)
upper_bound = np.percentile(bootstrap_proportions, (1 + confidence_level) / 2 * 100)

# Plotting the KDE of bootstrapped proportions
plt.figure(figsize=(10, 6))
sns.kdeplot(bootstrap_proportions, color='skyblue', fill=True, alpha=0.5)  # Changed color to skyblue
plt.axvline(x=observed_successes / n, color='red', linestyle='--', label='Observed Proportion (61.25%)')
plt.axvline(x=0.5, color='green', linestyle='--', label='Null Hypothesis Proportion (50%)')

# Confidence interval
plt.fill_betweenx(y=[0, max(sns.kdeplot(bootstrap_proportions, color='skyblue').get_lines()[0].get_ydata()) * 1.1], 
                   x1=lower_bound, x2=upper_bound, color='yellow', alpha=0.5, label='95% Confidence Interval')

# Labels and title
plt.title('KDE of Bootstrapped Proportions with Confidence Interval')
plt.xlabel('Proportion of Correct Identifications')
plt.ylabel('Density')
plt.legend()
plt.xlim(0, 1)
plt.ylim(0, max(sns.kdeplot(bootstrap_proportions, color='skyblue').get_lines()[0].get_ydata()) * 1.1)  # Extend y-axis for better visibility
plt.show()

# Output confidence intervals
print(f"95% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
```

```python
# now we will perform hypothesis testing
import numpy as np
from scipy import stats

# Setting seed for reproducibility
np.random.seed(130)

# Parameters
n = 80  # Sample size
x = 49  # Number of correct responses
p0 = 0.5  # Null hypothesis probability

# Conducting a one-sided binomial test
p_value = stats.binom_test(x, n, p=p0, alternative='greater')

print(f"P-value: {p_value}")
```

### Ethical Professionalism Considerations
    
Using p-values and hypothesis testing appropriately is an important ethical and professional responsibility of anyone doing data analysis. Actually, there is quite the quiet Contra-Versy (or is it Con-TROV-ersy?) around p-values. First, on a general level, it seems quite clear that p-values and hypothesis testing methodologies MUST play some ongoing contributing role in the so-called "replication crisis" rampantly afflicting mordern science; namely, "significant findings" made in scientific studies are not able to be reproduced by future studies at an alarming rate; and, this whole paradigm of "significant findings" is based on p-values and hypothesis testing... so, something's going on with this methodology in some way...
    
More specifically however, p-values are themselves quite problematic. To see this, just briefly consider the following article titles...

- [Why are p-values controversial?](https://www.tandfonline.com/doi/full/10.1080/00031305.2016.1277161) 
- [What a nerdy debate about p-values shows about science and how to fix it](https://www.vox.com/science-and-health/2017/7/31/16021654/p-values-statistical-significance-redefine-0005)
- [The reign of the p-value is over: what alternative analyses could we employ to fill the power vacuum?](https://royalsocietypublishing.org/doi/10.1098/rsbl.2019.0174)
- [Scientists rise up against statistical significance](https://www.nature.com/articles/d41586-019-00857-9)
- [Statistics experts urge scientists to rethink the p-value](https://www.spectrumnews.org/news/statistics-experts-urge-scientists-rethink-p-value)

While the issues here are relatively advanced and subtle (as introduced [here](https://www2.stat.duke.edu/~berger/p-values.html), presented [here](https://www.jarad.me/courses/stat587Eng/slides/Inference/I06-Pvalues/why_pvalues_dont_mean_what_you_think_they_mean.pdf), and demonstrated using simulation [here](https://jaradniemi.shinyapps.io/pvalue/)), the problem essentially comes down to the fact that most scientists (or just people) don't know how to really interpret the numeric value of a p-value. There are therefore two current proposed solutions to address this challenge.
    
1. Just interpreting p-values using the follwing table (which really isn't that hard, so it's surprising that this solution isn't more broadly adopted...)
    
|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|
    

2. Only do **hypothesis testing** on the basis of confidence intervals, not **p-values** (which might be the best solution wherever doing so is a realistic, convenient  possibility...)

With this quite broad introductory context in mind, what does your favorite ChatBot thinks about the following statements? 
    
1. Hypothesis testing is not a "mathematical proof"<br><br>

    1. We do not prove $H_0$ false, we instead give evidence against the $H_0$: "We reject the null hypothesis with a p-value of XYZ, meaning we have ABC evidence against the null hypothesis"
    2. We do not prove $H_0$ is true, we instead do not have evidence to reject $H_0$: "We fail to reject the null hypothesis with a p-value of XYZ"<br><br>

2. Implying that a "non-significant result" means there is "no effect" misleads an audience because this may in actual fact simply indicate that there was insufficient evidence to reject the null hypothesis. So this therefore overlooks the possibility of sample size limitations, or Type II errors (which means a test incorrectly concludes that there is no effect or difference when, in fact, there is one). 
    
> Similarly, analagously, a "significant result" used to reject the null hypothsis could alternatively be a Type I error (which means a test actually incorrectly rejected a null hypothesis when it was actually true)... we're only providing a measure of evidence against the null hypothesis... but the evidence could still incorrectly suggest the wrong conclusion... it really depends on how strong the evidence is...
>
> - all of which is why just interpreting p-values using the table above is a good idea...

3. The p-values used for hypothesis testing are contructed upone the assumptions of the null hypotheses they correspond to; but, null hypotheses are actually often presented in simple forms that routinely hide a lot of information that is implicitly used to construct the p-values. For example, distributional assumptions about the population, estimated "plug-in" values that can used to simplify the problem calculations, and the reliance upon "random sampling", etc...<br><br>
           
4. Drawing overly broad conclusions, or making recommendations based on findings that reject the null hypothesis in a specific context is fraught with the problematic risks of overgeneralization errors. Further exacerbating this issue, null hypotheses are typically so called "point null hypotheses" which is meant to emphasize that they are mathematically increadibly sharply specific; whereas, alternative hypotheses are usually very unspecific. An alternative hypothesis that "the null hypothesis is false" doesn't say much... we should wonder, "how, specfically, is the null false?"
    
As an example really giving a demonstrating this, consider rejecting a null hypothesis that there is no correlation between rain and pizza's delivered. Such a decision doesn't specify what the actual hypothetical correlation might be. In fact, it doesn't even indicate if there are more or less pizzas delivered when it rains... 

> which, actually, shows very clearly why statistical inference using hypothesis testing is inferior to statistical inference based on confidence intervals...
> 
> - a confidence interval provides a range of plausible values of what the parameter in question might be; whereas, ...
> - trying to more clearly address what the plausible values of the parameter in question might be on the basis of hypothesis testing would require conducting further experiements to continously reject increasingly detailed hypothesies to narrow down what the alternative hypothesis might actually include... which would indeed be an utterly vapid misuse of the intended purpose of hypothesis testing entrprise...
