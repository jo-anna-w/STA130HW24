```python
# https://github.com/veronicakwon/sta130h1/blob/main/cleaned_data%20(1).csv

import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/veronicakwon/sta130h1/refs/heads/main/cleaned_data%20(1).csv"
data = pd.read_csv(url)

print(data.head())  
print(data.info())  


plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='GEO_housing_live_with_dogs', y='LONELY_change_pre_covid') 


plt.title('Violin Plot of loneliness felt during COVID and dogs owned', fontsize=14)
plt.xlabel('number of dogs per household', fontsize=12)
plt.ylabel('how lonely people felt', fontsize=12)

plt.show()
```
